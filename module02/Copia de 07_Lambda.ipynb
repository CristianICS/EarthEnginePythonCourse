### **Funciones lambda**

Sirve para crear funciones anónimas. Ejemplo:


```
def double(x):
  return x**2
```

```
double(25)
```

```
double_lambda = lambda x: x**2
```

```
double_lambda(25)
```

### **Map**

Aplica una condición (definida por una función) sobretodos los elementos de una estructura de datos.
<img src="https://leblancfg.com/img/map_function.png" width="50%">



#### **Ejemplo 1**
Convertir nombre de mis mascotas a mayuscula

```
mascotas_nombre_min = ['alfred', 'tabitha', 'william', 'arla']
```

```
# Programación imperativa
mascotas_nombre_may = []
for pet in mascotas_nombre_min:
  pet_ = pet.upper()
  mascotas_nombre_may.append(pet_)
print(mascotas_nombre_may)
```

```
# Programacion funcional
list(map(lambda x:x.upper(), mascotas_nombre_min))
```

#### **Ejemplo 2**

Convertir los valores de temperatura media de Celsius a Kelvin

```
temp_Pucallpa_C = [28, 32, 15, 26, 24, 30]
```

```
# Programación imperativa
temp_Pucallpa_K = []
for temp in temp_Pucallpa_C:
  new_temp = temp + 273.15
  temp_Pucallpa_K.append(new_temp) 
print(temp_Pucallpa_K)   
```

```
list(map(lambda x: x + 273.15, temp_Pucallpa_C))  
```

### **Filter**

Mientras que **map()** pasa cada elemento del iterable a través de una función y devuelve el resultado de todos los elementos que han pasado a través de la función, **filter()**, en primer lugar, requiere que **la función devuelva valores booleanos (verdadero o falso)** y luego pasa cada elemento del iterable a través de la función, **"filtrando" los que son falsos**. 

<img src="https://leblancfg.com/img/filter.png" width="50%">

```
# Filter
temp_Pucallpa_C = [28, 32, 15, 26, 24, 30]
```

```
list(filter(lambda x:x<29, temp_Pucallpa_C))
```
